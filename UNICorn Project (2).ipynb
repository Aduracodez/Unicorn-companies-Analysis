```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#load and read the file
df = pd.read_csv("Unicorn_Companies.csv")
df
```

```python
#Check for the missing Values 
df.isnull().sum()
```

```python
#Drop the missing Values 
df.dropna(inplace=True)
df
```

```python
#To Clean the Valuation Column
df['Valuation'] = df['Valuation'].str.replace('$', '').str.replace('B', '').astype(int) * 1000000000
df
```

```python
#view the values in the funding columns
funding_counts = df['Funding'].value_counts()
#df = df[df['Funding'] != 'unknown']

print(funding_counts)
#print(df)
```

```python
df
```

```python
df2 = df.loc[df["Funding"] == "Unknown"]
df2
```

```python
df = df[df['Funding'] != 'unknown']

# Reset index if needed
df = df.reset_index(drop=True)

# Display the cleaned DataFrame
print(df)
```

```python
df
```

```python
df.drop(df[df['Funding'] == 'Unknown'].index, inplace=True)
df.reset_index(drop=True, inplace=True)
print(df)
```

```python
df
```

The Unknown Values has been dropped successfully which is further helpful for our Analysis now lets go ahead to clean the columns Further

```python
#lets clean the dollar sign and the letter "B&M"
df['Funding'] = df['Funding'].str.replace('$', '')

df['Funding'] = df['Funding'].apply(lambda x: float(x.replace('B', '')) * 1_000_000_000 if 'B' in x else float(x.replace('M', '')) * 1_000_000)
```

```python
#Convert to integer
df['Funding'] = df['Funding'].apply(lambda x: int(x))
```

```python
df
```

```python
df.dtypes
```

```python
#convert the date column to the right format
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
df
```

```python
#To check if it has been converted to the correct format
df.dtypes
```

```python
#To Clean the columns names 
df.rename(columns={'Date Joined': 'Date_Joined', 'Year Founded': 'Year_Founded', 'Select Investors': 'Select_Investors'}, inplace=True)
df
```

```python
#To Calculate the Return on Investment
df['ROI'] = round((df['Valuation'] - df['Funding']) / df['Funding'] * 100)
df
```

```python
df.dtypes
```

```python
# Find the index of the row with the highest ROI
highest_roi_company = df[df['ROI'] == df['ROI'].max()]

if not highest_roi_company.empty:
    print(f"The company with the highest ROI is {highest_roi_company['Company'].values[0]} with an ROI of {highest_roi_company['ROI'].values[0]}%.")
else:
    print("No data found.")
```

```python
max_roi_index = df['ROI'].idxmax()

# Display the entire row for the company with the highest ROI
highest_roi_row = df.loc[max_roi_index]
print("Company with the highest ROI:")
print(highest_roi_row)
```

```python
# Split the "Select_Investors" column into a list of investors
df['Investors'] = df['Select_Investors'].str.split(', ')

# Create a list of all investors
all_investors = [investor for investors_list in df['Investors'] for investor in investors_list]

# Count the frequency of each investor
investor_counts = pd.Series(all_investors).value_counts()

# Display the investors who have funded the most unicorn companies
print("Investors who have funded the most unicorn companies:")
print(investor_counts)
```

```python
# Sort the DataFrame by funding in ascending order
df_sorted = df.sort_values(by='Funding', ascending=True)

# Select the bottom 5 companies with the lowest funding
bottom_5_lowest_funding = df_sorted.head(5)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(bottom_5_lowest_funding['Company'], bottom_5_lowest_funding['Funding'], color='skyblue')
plt.title('Bottom 5 Companies with the Lowest Funding')
plt.xlabel('Funding Amount')
plt.ylabel('Company')

# Display actual funding values
for index, value in enumerate(bottom_5_lowest_funding['Funding']):
    plt.text(value, index, f'${value:.0f}', va='center')

plt.tight_layout()

# Show the plot
plt.show()








```

```python
# Group the data by city and calculate the average ROI for each city
city_avg_roi = df.groupby('City')['ROI'].mean().reset_index()

# Sort the cities by average ROI in descending order and select the top 5
top_5_cities = city_avg_roi.sort_values(by='ROI', ascending=False).head(5)

# Group the data by country and calculate the average ROI for each country
country_avg_roi = df.groupby('Country')['ROI'].mean().reset_index()

# Sort the countries by average ROI in descending order and select the top 5
top_5_countries = country_avg_roi.sort_values(by='ROI', ascending=False).head(5)

# Create bar charts for the top 5 cities and countries with the highest ROI
plt.figure(figsize=(12, 6))

# Bar chart for cities
plt.subplot(1, 2, 1)
plt.barh(top_5_cities['City'], top_5_cities['ROI'], color='skyblue')
plt.title('Top 5 Cities with the Highest ROI')
plt.xlabel('Average ROI')
plt.ylabel('City')

# Display data labels on the bars
for index, value in enumerate(top_5_cities['ROI']):
    plt.text(value, index, f'{value:.1f}', va='center')

# Bar chart for countries
plt.subplot(1, 2, 2)
plt.barh(top_5_countries['Country'], top_5_countries['ROI'], color='lightcoral')
plt.title('Top 5 Countries with the Highest ROI')
plt.xlabel('Average ROI')
plt.ylabel('Country')

# Display data labels on the bars
for index, value in enumerate(top_5_countries['ROI']):
    plt.text(value, index, f'{value:.1f}', va='center')

plt.tight_layout()
plt.show()
```

According to the  data it has shown that the the city with the highest ROI tends to be Sunnyvale wiht 57563 and the Country with the HIghest ROI tends to be Philippines  with 3444

```python
industry_avg_roi = df.groupby('Industry')['ROI'].mean().reset_index()

# Sort the industries by average ROI in descending order and select the top 5
top_5_industries = industry_avg_roi.sort_values(by='ROI', ascending=False).head(5)

# Create a bar chart to visualize the top 5 industries with the highest ROI
plt.figure(figsize=(10, 6))
plt.bar(top_5_industries['Industry'], top_5_industries['ROI'], color='skyblue')
plt.title('Top 5 Industries with the Highest ROI')
plt.xlabel('Industry')
plt.ylabel('Average ROI')

# Annotate the bars with the ROI values
for index, row in top_5_industries.iterrows():
    plt.text(row['Industry'], row['ROI'], f'{row["ROI"]:.1f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
